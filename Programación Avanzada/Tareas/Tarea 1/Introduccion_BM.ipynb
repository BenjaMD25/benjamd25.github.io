El siguiente código indica al usuario ingresar una lista de valores numéricos reales positivos. Cada valor es procesado como el radio de una circunferencia, a partir de estos procede a calcular el área y perímetro respectivo de cada una. En caso de ingresar un valor negativo o no numérico, el programa solicitará ingresar nuevamente la lista.

```python
from numpy import pi
while True:
    try:
        lista = list(input("Ingrese una lista solamente con números reales positivos separados por comas: ").split(","))
        for i in lista:
            a = float(i)
            if a<=0:
                lista.append('a')
        break
    except ValueError:
        print("Error: Debe ingresar una lista solamente con números reales positivos")
for i in lista:
    a = float(i)
    print('El área de la circunferencia con radio ', a, ' es ', a**2*pi)
    print('El perímetro de la circunferencia con radio ', a, ' es ', a*2*pi)
    print('________________________________________________________________')
```
