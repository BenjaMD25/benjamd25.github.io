Para el correcto funcionamiento de las celdas adjuntas en este archivo, se debe importar el módulo funciones.py

El siguiente código, por medio de las librerias matplotlib, numpy y el módulo funciones, realiza la gráfica de la función $f(x) = x^2$, con $x \in [-10,10]$.

```python
import matplotlib.pyplot as plt
import numpy as np
import funciones as fs

x = np.linspace(-10,10,100)
plt.plot(x,fs.f(x), color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.savefig('graficos/grafico_2d', bbox_inches='tight')
plt.show()
```

El siguiente código, por medio de las librerias plotly.graph, numpy y el módulo funciones, realiza la gráfica de la función $g(x) = \sin(x) \cdot \cos(y)$, con $x,y \in [-5,5]$.

```python
import plotly.graph_objects as go
from matplotlib.pyplot import savefig
from matplotlib import cm
import numpy as np
import funciones as fs

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)
Z = fs.g(X,Y)
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('g(x,y)')
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.title('Gráfica de la función g(x,y)')
savefig("graficos/grafico_3d")
plt.show()
```
